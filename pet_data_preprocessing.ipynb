```python
import pandas as pd
import matplotlib.pyplot as plt

#read in data 
chidat = pd.read_csv("chicago_adopted_20k_Jan_20.csv") 
stldat = pd.read_csv("StL_adopted_20k_Jan_21.csv")
inddat = pd.read_csv("Indy_adopted_20k_Jan_21.csv")

petdat = pd.concat([chidat, stldat, inddat])

#drop variables that aren't interesting out of the gate

petdat = petdat.drop(['Unnamed: 0',
                      'url',
                      'type',
                      'species',
                      'tags',
                      'description',
                      'attributes.declawed',
                      'contact.phone',
                      'contact.address.address1',
                      'contact.address.address2',
                      'contact.address.state',
                      'contact.address.postcode',
                      'contact.address.country',
                      'photos',
                      'colors.primary',
                      'colors.secondary',
                      'colors.tertiary',
                      '_links.self.href',
                      '_links.type.href'], axis=1)


#drop variables that have lots of NAs, leaving only features with no NAs (well, name has 1???)

petdat = petdat.drop(['breeds.secondary',
                      'environment.children',
                      'environment.dogs',
                      'environment.cats',
                      'contact.email',
                      'coat'], axis=1)


#reformat dates and calculate difference in dates (= target variable)

fmt ='%Y-%m-%dT%H:%M:%S+0000'
petdat['published_at'] =  pd.to_datetime(petdat['published_at'], format=fmt)
petdat['status_changed_at'] =  pd.to_datetime(petdat['status_changed_at'], format=fmt)
petdat['time_diff'] = petdat['status_changed_at'] - petdat['published_at']
petdat['target_hour'] = petdat.time_diff.astype('timedelta64[h]')

petdat.head()


#take out outliers 



```
