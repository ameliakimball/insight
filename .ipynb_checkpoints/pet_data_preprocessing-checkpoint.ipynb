```python
from datetime import datetime
fmt1 = '%Y-%m-%d %H:%M:%S'
fmt2 ='%Y-%m-%dT%H:%M:%S+0000'

tstamp1 = datetime.strptime('2016-04-07 09:06:02', fmt1)
tstamp2 = datetime.strptime('2020-01-19T23:05:58+0000', fmt2)

if tstamp1 > tstamp2:
    td = tstamp1 - tstamp2
else:
    td = tstamp2 - tstamp1
td_mins = int(round(td.total_seconds() / 60))
```
